```python
import torch # type: ignore
from diffusers import StableDiffusion3Pipeline # type: ignore

# 로컬에 저장된 모델 불러오기
save_path = "./stable-diffusion-3.5-model"
pipe = StableDiffusion3Pipeline.from_pretrained(save_path, torch_dtype=torch.bfloat16)

pipe = pipe.to("cuda")
```

```python

# 로컬에 저장된 모델 불러오기
save_path = "./stable-diffusion-3.5-model"
pipe = StableDiffusion3Pipeline.from_pretrained(save_path, torch_dtype=torch.bfloat16)

pipe = pipe.to("cuda")
```

```python
img_prompt = "coca cola logo"
img_name = "example"

image = pipe(
    img_prompt,
    num_inference_steps=40,
    guidance_scale=4.5,
).images[0]
image.save( img_name + ".png")
```

```python
import matplotlib.pyplot as plt  # type: ignore

plt.imshow(image)  # 이미지 콘솔에 출력
plt.axis('off')  # 축 숨기기
plt.show()  # 이미지 표시
```
